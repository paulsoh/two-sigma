```python
for col in df.columns:
    df[col].fillna(df[col].mean())
    
correlation_array = []
for col in df.columns:
        correlation_array.append(df[col].corr(df['y']))
correlation_array    
```

```python
labels = df.columns[2:-1]
values = correlation_array[2:-1]
```

```python
data = [go.Bar(
    y=labels,
    x=values,
    orientation= 'h'
)]
```
